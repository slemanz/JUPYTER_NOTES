## Generating a test signal

- The signal will consist in $y(t)$, where:

$$ y(t) = a \sin (2\pi \cdot f \cdot t) $$

In python:

```python
# Generate a signal
samplingFreq = 1000 # sampled at 1 kHz = 1000 samples / second
tlims = [0, 1]        # in seconds
signalFreq = 2 # Cycles / second
signalMag = 1 # magnitude of each sine
t = np.linspace(tlims[0],tlims[1],(tlims[1]-tlims[0])*samplingFreq)
y = signalMag*np.sin(2*math.pi*signalFreq*t)

# Plot the signal
plt.figure()
plt.plot(t,y)
plt.ylabel("$y(t)$")
plt.xlabel("$t$ (s)")
plt.xlim([min(t),max(t)])

plt.show()
```
